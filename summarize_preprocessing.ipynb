```python
import pandas as pd
import itertools
import json
import os
import glob
```

# Data Load

```python
def getter(data_type, path_ls=[]):
    original_sentence = []
    abstraction = []
    
    if data_type == '문서요약_텍스트':
        
        for path in path_ls:
            with open(path) as f:
                js = json.loads(f.read())
            df = pd.DataFrame(js)

            for i in range(len(js['documents'])):
                original_sentence_inner = []
                abstraction_inner = []
                row_data = df['documents'][i]
                row_text = row_data['text']
                for j in range(len(row_text)):
                    row_text_inner = row_text[j]
                    for k in range(len(row_text_inner)):
                        row_text_content = row_text_inner[k]['sentence']
                        if row_text_content[0] == '[':
                            row_text_content = row_text_content[1:-2]
                        original_sentence_inner.append(row_text_content)

                row_abstraction = df['documents'][i]['abstractive']
                abstraction_inner.append(row_abstraction[0])

                ','.join(original_sentence_inner)
                ','.join(abstraction_inner)

                original_sentence.append(original_sentence_inner)
                abstraction.append(abstraction_inner)
    
    elif data_type == '도서자료_요약' and path_ls:
        
        for path in path_ls:
            original_sentence_inner = []
            abstraction_inner = []
            with open(path) as f:
                js = json.loads(f.read())
            
            original_sentence_inner.append(js['passage'])
            abstraction_inner.append(js['summary'])
        
            original_sentence.append(original_sentence_inner)
            abstraction.append(abstraction_inner)
    
    elif data_type == '요약문_및_레포트_생성_데이터':
        for path in path_ls:
            original_sentence_inner = []
            abstraction_inner = []
            with open(path) as f:
                js = json.loads(f.read())
            
            original_sentence_inner.append(js['Meta(Refine)']['passage'])
            abstraction_inner.append(js['Annotation']['summary1'])
        
            original_sentence.append(original_sentence_inner)
            abstraction.append(abstraction_inner)
    
    elif data_type == '방송_콘텐츠_대본_요약_데이터':
        for path in path_ls:
            original_sentence_inner = []
            abstraction_inner = []
            with open(path) as f:
                js = json.loads(f.read())
            
            original_sentence_inner.append(js['Meta']['passage'])
            abstraction_inner.append(js['Annotation']['Summary1'])
        
            original_sentence.append(original_sentence_inner)
            abstraction.append(abstraction_inner)
    
    elif data_type == '논문자료_요약':
        for path in path_ls:
            original_sentence_inner = []
            abstraction_inner = []       
    
            with open(article_summ_train_files[0]) as f:
                js = json.loads(f.read())
                data = js['data']
                for i in range(len(data)):
                    original_sentence_inner = []
                    abstraction_inner = []
                    row_data = data[i]
                    
                    original_sentence_inner.append(row_data['summary_section'][0]['orginal_text'])
                    abstraction_inner.append(row_data['summary_section'][0]['summary_text'])

                    original_sentence.append(original_sentence_inner)
                    abstraction.append(abstraction_inner)
    
    df_summarize = pd.DataFrame({
        'original': original_sentence, 'abstraction': abstraction
    })
    df_summarize['original'] = df_summarize['original'].apply(lambda x : ','.join(x))
    df_summarize['abstraction'] = df_summarize['abstraction'].apply(lambda x : ','.join(x))
    
    return df_summarize
```

```python
path = 'data/문서요약_텍스트/Training/**/*.json'
file_list = glob.glob(path, recursive=True)
document_summ_train_files = [file for file in file_list if file.endswith('.json')]

path = 'data/문서요약_텍스트/Validation/**/*.json'
file_list = glob.glob(path, recursive=True)
document_summ_val_files = [file for file in file_list if file.endswith('.json')]

path = 'data/도서자료_요약/Training/도서요약_train/**/*.json'
file_list = glob.glob(path, recursive=True)
book_summ_train_files = [file for file in file_list if file.endswith('.json')]

path = 'data/도서자료_요약/Training/도서요약_valid/**/*.json'
file_list = glob.glob(path, recursive=True)
book_summ_val_files = [file for file in file_list if file.endswith('.json')]

path = 'data/022_요약문_및_레포트_생성_데이터/1.Training/라벨링데이터/**/*.json'
file_list = glob.glob(path, recursive=True)
report_summ_train_files = [file for file in file_list if file.endswith('.json')]

path = 'data/022_요약문_및_레포트_생성_데이터/2.Validation/라벨링데이터/**/*.json'
file_list = glob.glob(path, recursive=True)
report_summ_val_files = [file for file in file_list if file.endswith('.json')]

path = 'data/023_방송_콘텐츠_대본_요약_데이터/01.데이터/1.Training/라벨링데이터/**/*.json'
file_list = glob.glob(path, recursive=True)
broadcast_summ_train_files = [file for file in file_list if file.endswith('.json')]

path = 'data/023_방송_콘텐츠_대본_요약_데이터/01.데이터//2.Validation/라벨링데이터/**/*.json'
file_list = glob.glob(path, recursive=True)
broadcast_summ_val_files = [file for file in file_list if file.endswith('.json')]

path = 'data/논문자료_요약/Training/training_논문/**/*.json'
file_list = glob.glob(path, recursive=True)
article_summ_train_files = [file for file in file_list if file.endswith('.json')]

path = 'data/논문자료_요약/Validation/validation_논문/**/*.json'
file_list = glob.glob(path, recursive=True)
article_summ_val_files = [file for file in file_list if file.endswith('.json')]
```

```python
df_train_data_docs = getter('문서요약_텍스트', document_summ_train_files)
df_val_data_docs = getter('문서요약_텍스트', document_summ_val_files)
```

```python
df_train_data_book = getter('도서자료_요약', book_summ_train_files)
df_val_data_book = getter('도서자료_요약', book_summ_val_files)
```

```python
df_train_data_report = getter('요약문_및_레포트_생성_데이터', report_summ_train_files)
df_val_data_report = getter('요약문_및_레포트_생성_데이터', report_summ_val_files)
```

```python
df_train_data_broadcast = getter('방송_콘텐츠_대본_요약_데이터', broadcast_summ_train_files)
df_val_data_broadcast = getter('방송_콘텐츠_대본_요약_데이터', broadcast_summ_val_files)
```

```python
df_train_data_article = getter('논문자료_요약', article_summ_train_files)
df_val_data_article = getter('논문자료_요약', article_summ_val_files)
```

# Integration

```python
df_train = pd.concat([df_train_data_docs, df_train_data_book, df_train_data_report, df_train_data_article], axis=0)
```

```python
df_test = pd.concat([df_val_data_docs, df_val_data_book, df_val_data_report, df_val_data_article], axis=0)
```
